```python
3)Explain the following terms
 The difference in np.array, np.asarray and np.asanyarray
 The difference between Deep copy and shallow copy
np.array:

np.array creates a new array object from any object exposing the array interface (like lists or tuples).
It always copies the data, even if the input object is already an array.
np.asarray:
np.asarray also creates an array, but it will not copy the input if the input is already an ndarray. 
It will simply return the original array object if the input is already an ndarray.
If the input is not an array, it behaves like np.array and creates a new array.
This is more efficient because it avoids unnecessary copying.
    np.asanyarray:

np.asanyarray is similar to np.asarray but allows subclasses of ndarray (such as a matrix or a subclass) to be preserved in the result.
It doesn't convert subclass objects into a basic ndarray; it will return the input if it's already an instance of a subclass of ndarray.
Shallow Copy:

A shallow copy creates a new object but does not recursively copy the objects within it. Instead, it copies references to the objects inside.
As a result, if you modify an object inside the shallow copy, 
it can affect the original object, as both the original and copied objects share the same references to the inner objects.
Deep Copy:

A deep copy creates a new object and recursively copies all objects inside it. 
The new object is entirely independent of the original, meaning that changes made to the deep copy do not affect the original.
```

```python
5) Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array 
perform the following operations:
 a)Extract all even integers from array.
 b)Extract all odd integers from array
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
array = np.random.randint(1, 11, size=(5, 6))

print("Original array:")
print(array)

# a) Extract all even integers
even_numbers = array[array % 2 == 0]

# b) Extract all odd integers
odd_numbers = array[array % 2 != 0]

print("\nEven integers:")
print(even_numbers)

print("\nOdd integers:")
print(odd_numbers)
```

```python
18)With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine 
Wave Function'
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x-values (from 0 to 2*pi)
x = np.linspace(0, 2 * np.pi, 100)

# Compute y-values using sine function
y = np.sin(x)

# Create a new plot (figure)
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)', tools="pan,box_zoom,reset")

# Add line plot
p.line(x, y, legend_label="Sine Wave", line_width=2, color="blue")

# Show grid lines
p.grid.grid_line_alpha = 0.3

# Display the plot in the notebook
output_notebook()  # If
```

```python
19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their 
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical 
Bar Chart'
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral11

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 20, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a new figure
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           x_axis_label='Category', y_axis_label='Value', tools="pan,box_zoom,reset")

# Create a bar chart
p.vbar(x='categories', top='values', width=0.9, source=source, 
       legend_field="categories", color=Spectral11[:len(categories)])

# Add hover tool to show exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Display the grid lines and make the legend visible
p.grid.grid_line_alpha = 0.3
p.legend.location = "top_left"

# Display the plot in the notebook
output_notebook
```
